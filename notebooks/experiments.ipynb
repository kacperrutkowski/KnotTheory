```python
import sys
from pathlib import Path
```

```python
sys.path.append(str(Path().resolve().parent))
```

```python
import src.knots as kn
import pandas as pd
```

```python
DATA_PATH = Path().resolve().parent / "data"

df_homfly = pd.read_csv(DATA_PATH / "homfly.csv")
df_jones = pd.read_csv(DATA_PATH / "jones.csv")
df_kauffman = pd.read_csv(DATA_PATH / "kauffman.csv")
```

Zamiana napisów przedstawiajacych wektor na listy:

```python
df_jones["Jones (vector)"] = df_jones["Jones (vector)"].apply(lambda x: kn.string_to_vector_parser(x))
```

```python
df_homfly["HOMFLY (vector)"] = df_homfly["HOMFLY (vector)"].apply(lambda x: kn.string_to_vector_parser(x))
```

```python
df_kauffman["Kauffman (vector)"] = df_kauffman["Kauffman (vector)"].apply(lambda x: kn.string_to_vector_parser(x))
```

Dodanie kolumny z wartościami True lub False w zależności od tego czy dany węzeł jest symetryczny, czy nie:

```python
df_jones["is_symmetrical"] = df_jones["Jones (vector)"].apply(lambda x: kn.is_symmetrical(x))
```

```python
df_homfly["is_symmetrical"] = df_homfly["HOMFLY (vector)"].apply(lambda x: kn.is_symmetrical(x, vector_type = "homfly"))
```

```python
df_kauffman["is_symmetrical"] = df_kauffman["Kauffman (vector)"].apply(lambda x: kn.is_symmetrical(x, vector_type = "kauffman"))
```

Dodanie kolumny z wartościami True lub False w zależności od tego czy węzeł jest amphicheiral czy nie: 

```python
df_jones["is_amphicheiral"] = df_jones["Symmetry Type"].str.contains("amphicheiral")
```

```python
df_homfly["is_amphicheiral"] = df_homfly["Symmetry Type"].str.contains("amphicheiral")
```

```python
df_kauffman["is_amphicheiral"] = df_kauffman["Symmetry Type"].str.contains("amphicheiral")
```

Wyniki:

```python
print(f"Liczba węzłów rozpoznanych jako symetryczne przez wielomian jones'a: {sum(df_jones['is_symmetrical'])}")
print(f"Liczba węzłów niepoprawnie rozpoznanych jako symetryczne przez wielomian Jonesa: {(df_jones['is_symmetrical'] & ~df_jones['is_amphicheiral']).sum()}")
```

```python
print(f"Liczba węzłów rozpoznanych jako symetryczne przez wielomian HOMFLY: {sum(df_homfly['is_symmetrical'])}")
print(f"Liczba węzłów niepoprawnie rozpoznanych jako symetryczne przez wielomian HOMFLY: {(df_homfly['is_symmetrical'] & ~df_homfly['is_amphicheiral']).sum()}")
```

```python
print(f"Liczba węzłów rozpoznanych jako symetryczne przez wielomian kauffman'a: {sum(df_kauffman['is_symmetrical'])}")
print(f"Liczba węzłów niepoprawnie rozpoznanych jako symetryczne przez wielomian Kauffmana: {(df_kauffman['is_symmetrical'] & ~df_kauffman['is_amphicheiral']).sum()}")
```
